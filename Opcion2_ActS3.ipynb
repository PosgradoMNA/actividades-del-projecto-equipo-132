```python
#Conceptos basicos (Insights)

#Base de datos y Data Warehouse

#Una base de datos es un conjunto de informacion la cual esta organizada normalmente en forma de tabla (columnas, filas, celdas), cada celda contiene los datos de cada registro, usualmente
#la primer fila de la tabla esta reservado para el nombre de las columnas/variables de la base de datos.
#Un almacen de datos es un conjunto de datos que se recopilan de diferentes fuentes y se acomodan de una manera logica y entendible para quien vaya a usar la informacion.
#Las dos principales formas de hacer un almacen de datos es con ETL o ELT, ETL es la forma que era usada previamente al primero extraer la informacion, luego transformarla fuera de linea 
#y posteriormente se carga en el almacen de datos, esto lo hace menos eficiente con la tecnologia actual, donde ELT es una forma mas usada utilizando las grandes capacidades de los
#almacenes de datos donde se puede cargar la informacion como se tiene, y ahi mismo se hace la transformacion, haciendo lo mas eficiente para bases de datos muy grandes.


#Pre-procesamiento de la información

#En ocasiones es necesario realizar un pre-procesamiento de datos, los cuales incluye limpieza de los datos que hay en una base de datos, con la finalidad de mejorar la calidad de la 
#informacion para realizar un analisis mas acertado, parte de esta limpieza involucra eliminar valores nulos de algunas celdas o filas, una opcion alterna es rellenar esos valores nulos 
#con algun dato especifico, ya sea un dato fijo o si hay valores numericos utilizar la media, mediana o cualquier otro valor que pudiera ayudar a asignar un valor, es parte tambien
#de la limpieza de datos el identificar valores atipicos o que pudieran meter ruido al analisis.
#Adicionalmente el procesamiento de datos involucra la normalizacion de los datos y en casos muy especificos reducir la cantidad de datos para facilitar el analisis de la informacion.
```
